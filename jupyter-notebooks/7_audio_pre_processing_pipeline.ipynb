This notebook demonstrates the pre-processing pipeline that is done to the nsynth dataset.
For the time being no data augmentation is done as the dataset seems comprehensive enough for the time being.

```python
import librosa
from librosa.effects import trim
import numpy as np
from typing import List
from utils.audio_processors import segment_signal
import utils.constants as consts
from utils.displays import display_wave_form
from utils.displays import display_audio_player
```

```python
# Load audio.s
keyboard_sample: np.ndarray = librosa.load('../media/audio/keyboard.wav', consts.SAMPLE_RATE, mono=True)[0]
```

##### Pre-onset trimming waveform.

```python
display_wave_form(keyboard_sample)
```

##### Trimmed waveform.

```python
trimmed_sample, _ = trim(keyboard_sample, top_db=consts.TRIM_DB)

display_wave_form(trimmed_sample)
```

##### Segmenting signal into 1 second segments size of global SAMPLE_RATE.

End of signal in this case has zero padding.

```python
segmented_sample: List[np.ndarray] = segment_signal(trimmed_sample)

# Display second segment.
display_wave_form(segmented_sample[1])
display_audio_player(segmented_sample[1])
```
